## CNN

```python
import pickle
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Conv2D,MaxPooling2D,MaxPool1D, Dense,Flatten,Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import LeakyReLU
```

```python
class CNNModel(tf.keras.Model):
    
    def __init__(self):
        super(CNNModel, self).__init__()
   
        self.conv1 = Conv1D(filters=32, kernel_size=5, padding='same', activation=LeakyReLU(alpha=0.001), input_shape = (205, 1))
        self.conv2 = Conv1D(filters=64, kernel_size=5, padding='same', activation=LeakyReLU(alpha=0.001))
        self.conv3 = Conv1D(filters=128, kernel_size=5, padding='same', activation=LeakyReLU(alpha=0.001))

        self.max_pool1 = MaxPool1D(pool_size=5, strides=2, padding='same')

        self.dropout = Dropout(0.5)

        self.flatten = Flatten()
        
        self.fcl = Dense(units=512, activation=LeakyReLU(alpha=0.001))
        self.fc2 = Dense(units=4, activation='softmax')


    # forward
    def call(self, x):
        
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.max_pool1(x)
        x = self.dropout(x)
        x = self.flatten(x)
        
        # full layer connection 
        x = self.fcl(x)
        x = self.fc2(x)
        return x
        
       
model = CNNModel()

model.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['mae'])
train_x = train.drop(['id','label'], axis=1)
train_x
```

```python
train_x = train_x.values.reshape(train_x.shape[0], 205, 1)
train_x
```

```python
train_y = tf.keras.utils.to_categorical(train['label'],4)
train_y
```

```python
model.fit(train_x, train_y, batch_size=128, epochs=1, verbose=1)
```

```python
test_x = test.drop(['id'], axis=1)
test_x = test_x.values.reshape(test_x.shape[0], 205, 1)
```

```python
y_pred = model.predict(test_x, batch_size=128)
y_pred
```
